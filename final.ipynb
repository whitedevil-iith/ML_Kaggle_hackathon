```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
import matplotlib.pyplot as plt


from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from scipy.stats import pearsonr, spearmanr


```

```python
# Replace 'your_file.csv' with the path to your CSV file
train_path = 'train.csv'
test_path = 'test.csv'

# Read the CSV file into a DataFrame
train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

copy_train = train_data
copy_test = test_data
# # Display the first few rows of the data
# print(train_data.head(5))
# print(test_data.head(5))

print(train_data.columns)
data = train_data
```

```python
threshold = 0.6  # 80% missing values

# Calculate the threshold for non-null values (1 - threshold)
threshold_count = int((1 - threshold) * len(data))

# Drop columns where the number of non-null values is less than the threshold count
data = data.dropna(axis=1, thresh=threshold_count)

# Display the cleaned DataFrame
print(data.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif
from scipy.stats import pearsonr, spearmanr

# Initialize the label encoder
label_encoder = LabelEncoder()

# Fit and transform the target column
data['Target'] = label_encoder.fit_transform(data['Target'])

# Filter out only numeric columns for analysis
numeric_data = data.select_dtypes(include='number')

# Iterate over each numeric column and calculate correlations
for column in numeric_data.columns:
    if column != 'Target':  # Skip the target column itself
        print(f"Calculating for {column}...")

        # Calculate Pearson correlation
        try:
            pearson_corr, _ = pearsonr(numeric_data[column], numeric_data['Target'])
            print(f'\tPearson Correlation for {column}: {pearson_corr}')
        except Exception as e:
            print(f"\tError calculating Pearson correlation for {column}: {e}")

        # Calculate Spearman correlation
        try:
            spearman_corr, _ = spearmanr(numeric_data[column], numeric_data['Target'])
            print(f'\tSpearman Correlation for {column}: {spearman_corr}')
        except Exception as e:
            print(f"\tError calculating Spearman correlation for {column}: {e}")
        
        # For classification problems
        try:
            mi = mutual_info_classif(data[[column]], data['Target'])
            print(f'\tMutual Information (Classification) for {column}: {mi[0]}')
        except Exception as e:
            print(f"\tError calculating Mutual Information for {column}: {e}")

        # Calculate the correlation matrix
        try:
            correlation_matrix = data.corr()
            # Get the correlation of a specific feature with the target
            feature_target_correlation = correlation_matrix['Target'][column]
            print(f'\tFeature-Target Correlation for {column}: {feature_target_correlation}')
        except Exception as e:
            print(f"\tError calculating feature-target correlation for {column}: {e}")
```

```python
# Define a threshold for feature-target correlation
correlation_threshold = 0.04  # You can change this value as needed

# Filter out only numeric columns for analysis
numeric_data = data.select_dtypes(include='number')

remove_feature_target_correlation_columns = []

# Iterate over each numeric column and calculate correlations
for column in numeric_data.columns:
    if column != 'Target':  # Skip the target column itself
        print(f"Calculating for {column}...")
        # Calculate the correlation matrix
        try:
            correlation_matrix = data.corr()
            # Get the correlation of a specific feature with the target
            feature_target_correlation = correlation_matrix['Target'][column]
            print(f'\tFeature-Target Correlation for {column}: {feature_target_correlation}')
            
            # Check if the absolute correlation is below the threshold
            if abs(feature_target_correlation) < correlation_threshold:
                print(f"\t\tFeature '{column}' has a low correlation with the target.")
                remove_feature_target_correlation_columns.append(column)
        except Exception as e:
            print(f"\tError calculating feature-target correlation for {column}: {e}")
```

```python
print(remove_feature_target_correlation_columns)
```

```python
mode_values = {}
# Function to replace missing values (NaN) with mode of the column
def replace_missing_with_mode(df):
    for column in df.columns:
        # If the column has missing values, replace them with the mode (most frequent value)
        if df[column].isnull().any():
            mode_value = df[column].mode()[0]  # Get the most frequent value
            mode_values[column] = mode_value
            # print(column," : ", mode  _value)
            df[column].fillna(mode_value, inplace=True)  # Replace NaN with the mode
    return df

# Replace missing values with the mode of each column
data = replace_missing_with_mode(data)


# Drop columns by specifying their names
columns_to_drop = remove_feature_target_correlation_columns
data = data.drop(columns=columns_to_drop)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split, GridSearchCV
import xgboost as xgb

# Prepare the data
testInput = test_data
trainInput = data.drop('Target', axis=1)
trainOutput = data['Target']

# Split the data into train and validation sets with stratification
X_train, X_val, y_train, y_val = train_test_split(
    trainInput, trainOutput, 
    test_size=0.2,  # 20% of the data for validation
    stratify=trainOutput,     # Stratify based on the target variable
    random_state=42 # Setting a random state for reproducibility
)

# Set up parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [100],  # Number of trees
    'learning_rate': [0.01],  # Step size shrinkage
    'max_depth': [5],  # Maximum tree depth
    'subsample': [0.8],  # Fraction of samples to use for training each tree
    'colsample_bytree': [0.8]  # Fraction of features to use for each tree
}

# Initialize XGBoost classifier
xgboost_classifier = xgb.XGBClassifier(random_state=42, eval_metric='mlogloss')

# Set up GridSearchCV with multi-class classification
grid_search = GridSearchCV(xgboost_classifier, param_grid, scoring='accuracy', cv=5, n_jobs=-2)
grid_search.fit(X_train, y_train)

# Best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Predictions on the validation set
predictions_XGB = best_model.predict(X_val)

# Accuracy, Precision, Recall (Macro-averaged for multi-class), and F1 Score
accuracy_xgb = accuracy_score(y_val, predictions_XGB)
precision_xgb = precision_score(y_val, predictions_XGB, average='macro')  # Macro-average precision
recall_xgb = recall_score(y_val, predictions_XGB, average='macro')  # Macro-average recall
f1_xgb = f1_score(y_val, predictions_XGB, average='macro')  # Macro-average F1 score

# Print results
print("Best XGBoost Parameters:", best_params)
print("XGBoost Accuracy:", accuracy_xgb)
print("XGBoost Precision (Macro):", precision_xgb)
print("XGBoost Recall (Macro):", recall_xgb)
print("XGBoost F1 Score (Macro):", f1_xgb)
```
